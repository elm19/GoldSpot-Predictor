```python
import pandas as pd
import pandas_datareader as pdr
import yfinance as yf
from datetime import datetime
import os

# Set the FRED API key in the environment
api_key = "1c5acb95620e0ed0d8052f1852190fbf"

# Configuration
start_date = datetime(2010, 1, 1)

#Taux de la Fed
fed_rate = pdr.get_data_fred("FEDFUNDS", start_date)

# DXY Index
dxy = yf.download("DX-Y.NYB", start=start_date)

# CPI
cpi = pdr.get_data_fred("CPIAUCSL", start_date)

# TIPS
tips = pdr.get_data_fred("DFII10", start_date)

# Taux de chômage américain (UNRATE)
unemployment = pdr.get_data_fred("UNRATE", start=datetime(2000, 1, 1))
```

```python
# Combine data
combined_data = pd.concat([fed_rate, cpi, tips,unemployment,dxy], axis=1)
combined_data
```

```python
print('nombre de ligne     :',combined_data.shape[0])
missing_values=combined_data.isnull().sum()
print(missing_values)
```

```python
combined_data['FEDFUNDS'] = combined_data["FEDFUNDS"].fillna(method='ffill')
combined_data['CPIAUCSL'] = combined_data["CPIAUCSL"].fillna(method='ffill')
macro_cols = ['FEDFUNDS', 'CPIAUCSL']
combined_data[macro_cols] = combined_data[macro_cols].bfill()

combined_data['UNRATE'] = combined_data["UNRATE"].fillna(method='ffill')


combined_data['DFII10'] = combined_data['DFII10'].interpolate(method='time').ffill().bfill()
```

```python
print(combined_data.columns)
```

```python
new_names = {
    ('Close', 'DX-Y.NYB'): 'Close_DX',
    ('High', 'DX-Y.NYB'): 'High_DX',
    ('Low', 'DX-Y.NYB'): 'Low_DX',
    ('Open', 'DX-Y.NYB'): 'Open_DX',
    ('Volume', 'DX-Y.NYB'): 'Volume_DX'
}

combined_data = combined_data.rename(columns=new_names)
dollar_cols = ['Close_DX', 'High_DX', 'Low_DX', 'Open_DX']
combined_data[dollar_cols] = combined_data[dollar_cols].interpolate(method='linear').ffill().bfill()

combined_data['Volume_DX'] = combined_data['Volume_DX'].fillna(0)
```

```python
print('nombre de ligne     :',combined_data.shape[0])
missing_values=combined_data.isnull().sum()
print(missing_values)
```
