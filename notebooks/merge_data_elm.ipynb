```python
import pandas as pd
import numpy as np

def custom_date_parser(date_str):
    return pd.to_datetime(date_str, format='%Y-%m-%d')
```

```python
callender=pd.read_csv('../data/processed-data/economic_calendar_19_24.csv', index_col='Date',  parse_dates=True)
price_data =pd.read_csv('../data/processed-data/price-gold.csv', index_col='Date',  parse_dates=True)
indicators = pd.read_csv('../data/processed-data/macro_data_process.csv', index_col='Date',  parse_dates=True)

print("Shape of callender dataframe:", callender.shape)
print("Shape of price_data dataframe:", price_data.shape)
print("Shape of indicators dataframe:", indicators.shape)
price_data
```

```python
# Data Merging and Analysis

In the following slides, we will perform data merging and analysis tasks. This includes combining datasets, filling missing
values, and preparing the data for further analysis and visualization.
```

```python
df = price_data.asfreq("D").ffill().reset_index().set_index('Date').loc[callender.index.min():callender.index.max()]
df_merged = pd.merge(df, callender, how='left', left_index=True, right_index=True)
```

```python
df_merged.fillna(0, inplace=True)
df_merged.head()
df_merged.to_csv('../data/processed-data/meged_data_price_callender.csv')
```

```python
df_merged2 = pd.merge(df, indicators, how='left', left_index=True, right_index=True)
df_merged2.fillna(method='ffill', inplace=True) 
df_merged.to_csv('../data/processed-data/meged_data_price_indicators.csv')
df_merged2.head()
```

```python
df_merged3 = pd.merge(df_merged, indicators, how='left', left_index=True, right_index=True)
df_merged3.fillna(method='ffill', inplace=True) 
df_merged.to_csv('../data/processed-data/meged_data_all.csv')
df_merged3.head()
```

# Data Reduction Plan

The datasets in this notebook contain a large number of columns compared to the number of rows. To make the data more manageable and improve the efficiency of analysis, we will reduce the number of columns by identifying and removing redundant or irrelevant features. Below are the methods we plan to use for this reduction:

1. **Correlation Analysis**: Identify highly correlated columns and remove one of the correlated pairs to avoid redundancy.
2. **Feature Importance**: Use machine learning models (e.g., Random Forest or Gradient Boosting) to determine the most important features and drop the less significant ones.
3. **Variance Threshold**: Remove columns with low variance as they contribute little to the analysis.
4. **Domain Knowledge**: Leverage domain expertise to identify and drop columns that are irrelevant to the analysis.
5. **Principal Component Analysis (PCA)**: Apply PCA to reduce dimensionality while retaining most of the variance in the data.
